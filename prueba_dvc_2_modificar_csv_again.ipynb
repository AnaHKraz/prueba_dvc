```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import array

pd.set_option('display.max_columns', None)
```

```python
ffile = 'C:/Users/anahe/Documents/prueba_dvc/data/full_dataset.csv' 
df = pd.read_csv(ffile, sep=",")
df
```

```python
# Calcular el precio por unidad
df['VALORBASEUNIDAD'] = df['VALORBASE'] / df['UNIDADES']
df['VALORNETOUNIDAD'] = df['VALORNETO'] / df['UNIDADES']

df.loc[df['VALORBASE'] == 0,'VALORBASEUNIDAD'] = 0
df.loc[df['VALORNETO'] == 0,'VALORNETOUNIDAD'] = 0
```

```python
# Unidades totales del pedido anterior
df_totalArticulos = df.groupby('CODPEDIDO')['UNIDADES'].sum().reset_index().rename(columns={'UNIDADES':'UNIDADESTOTALESANTERIOR'})
df = df.merge(df_totalArticulos[['CODPEDIDO','UNIDADESTOTALESANTERIOR']], on='CODPEDIDO')

# Contar el número de artículos diferentes en el pedido anterior
df_articuloDiferentes = df.groupby(['CODONEKEY','CODPEDIDO'])['CODARTICULOSAP'].nunique().shift().reset_index().rename(columns={'CODARTICULOSAP':'ARTICULOSDIFERENTESANTERIOR'})
df = df.merge(df_articuloDiferentes[['CODPEDIDO','ARTICULOSDIFERENTESANTERIOR']], on='CODPEDIDO')

df.sort_values(['FECHAFACTURA', 'CODPEDIDO'], inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
df.drop(columns=['VALORBASE','VALORNETO','CODFAMILIA'], inplace=True)
```
